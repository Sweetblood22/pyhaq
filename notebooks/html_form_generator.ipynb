```python
question_template = """
<div class="question-container" data-answered="false" data-value="0">
    <div class="question">{question_text}</div>
    <div class="answer-container">
        {response_field}
    </div>
</div>"""

response_template = """<div class="answer" data-value="{value}" onclick="selectAnswerExclusive(this)">{text}</div>"""

def generate_quiz(questions, answers_values, header=None, title=None):
    qtml = ""
    for i, question_text in enumerate(questions):
        response_field = '\n'.join((response_template.format(value=value, text=text) for text, value in answers_values)).strip()
        qtml = qtml + '\n' + question_template.format(question_text=question_text, response_field=response_field)
        
    return qtml.strip()
```

```python
def make_header(header, inner_element='p', title=None):
    title_text = f'\n  <div class="question header-title">{title}</div>' if title is not None else ''
    header_template = """
<div class="question-container header">{title_text}
  <div class="answer-container header">
    {head_labels}
  </div>
</div>"""
    head_template = """<div class="answer header">{text}</div>"""
    head_labels = '\n    '.join((head_template.format(text=head) for head in header)).strip()
    return header_template.format(title_text=title_text, head_labels=head_labels)
```

```python
def generate_table_responses(questions, header, values):
    qtml = ""
    assert len(header) == len(values)
    
    for i, question_text in enumerate(questions):
        response_field = '\n'.join((response_template.format(value=value, text=' ') for value in values)).strip()
        qtml = qtml + '\n' + question_template.format(question_text=question_text, response_field=response_field)
        
    return qtml.strip()
```

```python
def make_callback(n_questions, incomplete_warn, present_result, identifier):
    return """<script>
    function calculateScore() {
      let totalScore = 0;
      let unanswered = false;

      // Loop through the %d questions
      for (let i = 1; i <= %d; i++) {
        const radios = document.getElementsByName('%s' + i);
        let answered = false;

        for (let j = 0; j < radios.length; j++) {
          if (radios[j].checked) {
            totalScore += parseInt(radios[j].value);
            answered = true;
            break;
          }
        }

        if (!answered) {
          unanswered = true;
        }
      }

      if (unanswered) {
        document.getElementById('result').textContent = "%s";
      } else {
        document.getElementById('result').textContent = "%s" + totalScore;
      }
    }
  </script>""" %(n_questions, n_questions, identifier, incomplete_warn, present_result)

incomplete_warn, present_result = "모든 질문에 답변해 주세요.", "총 점수는 다음과 같습니다: "

print(make_callback(len(gad_burn_text), incomplete_warn, present_result, 'q'))
```
